```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Предположим, что data1 и data2 уже загружены и имеют общие столбцы для объединения (например, 'ID')

# 1. Горизонтальное объединение (по ключу)
data_combined = pd.merge(data1, data2, how='left', on='Id')

# 2. Фильтрация только числовых столбцов
numeric_columns = data_combined.select_dtypes(include=['float64', 'int64']).columns

# 3. Выводим гистограмму или плотность распределения для числовых столбцов (можно ограничить количество)
for column in numeric_columns[:10]:  # Например, выводим только первые 10 числовых столбцов
    plt.figure(figsize=(10, 6))
    sns.histplot(data_combined[column], kde=True)  # kde=True добавит кривую плотности
    plt.title(f'Гистограмма и плотность для {column}')
    plt.xlabel(column)
    plt.ylabel('Частота')
    plt.tight_layout()
    plt.show()

# Если данные очень большие, можно выбрать случайную подвыборку для анализа
# Пример: отбор случайных 1000 строк из объединенного набора данных
data_sample = data_combined.sample(n=1000, random_state=42)

# Выводим графики для числовых столбцов из подвыборки
for column in numeric_columns[:10]:  # Только первые 10 столбцов
    plt.figure(figsize=(10, 6))
    sns.histplot(data_sample[column], kde=True)  # kde=True добавит кривую плотности
    plt.title(f'Гистограмма и плотность для {column} (сэмпл)')
    plt.xlabel(column)
    plt.ylabel('Частота')
    plt.tight_layout()
    plt.show()
```

```python
# 1. Проверим, сколько уникальных Order_id есть в data_123 (до объединения)
unique_ids_main = data_123['Order_id'].nunique()

# 2. Сколько из них есть в таблице заказов
matching_ids = data_123['Order_id'].isin(orders_data['Order_id']).sum()

# 3. Сколько записей получили заказ (не NaN после объединения)
joined_orders_count = data_with_orders['order_date_time'].notnull().sum()

# 4. Сравнение
print(f"Уникальных Order_id в обращениях: {unique_ids_main}")
print(f"Сколько из них нашли совпадения в заказах: {matching_ids}")
print(f"Сколько строк после объединения получили данные по заказам: {joined_orders_count}")

# 5. Быстрая проверка конкретных NaN: убедимся, что в таблице заказов реально нет этих ID
missing_ids = data_with_orders[data_with_orders['order_date_time'].isnull()]['Order_id'].dropna().unique()
# Проверим первые 5
print("\nПервые 5 Order_id без данных заказа, ищем их в таблице заказов:")
for oid in missing_ids[:5]:
    found = orders_data[orders_data['Order_id'] == oid]
    print(f"{oid}: {'НАЙДЕН' if not found.empty else 'не найден'}")

# 6. Тепловая карта пропусков — визуальная проверка
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
sns.heatmap(data_with_orders.isnull(), yticklabels=False, cbar=False)
plt.title("Распределение пропущенных значений после объединения")
plt.show()
```

```python
join_types = ['inner', 'left', 'right', 'outer']
results = {}

for how in join_types:
    merged = data_123.merge(orders_data, on='Order_id', how=how)
    results[how] = {
        'rows': merged.shape[0],
        'cols': merged.shape[1],
        'missing_order_data': merged['order_date_time'].isnull().sum(),
    }

# Вывод результатов
print("Результаты по типам объединения:")
for join_type, stats in results.items():
    print(f"\n{join_type.upper()} JOIN:")
    print(f"Строк: {stats['rows']}")
    print(f"Столбцов: {stats['cols']}")
    print(f"Пропущенных значений в 'order_date_time': {stats['missing_order_data']}")
```
