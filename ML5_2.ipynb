```python
bin_ranges = [0, 250, 320, 10000]
bin_names = [1, 2, 3]
prsa_data['CO_bin_custom_label'] = pd.cut(np.array(prsa_data['CO']),
                                               bins=bin_ranges, labels=bin_names)
prsa_data.head()
```

```python
prsa_data['CO_bin_custom_label'] = prsa_data['CO_bin_custom_label'].values.add_categories(0)
prsa_data['CO_bin_custom_label'] = prsa_data['CO_bin_custom_label'].fillna(0).astype(int)
prsa_data.head()
```

```python
prsa_data.PRES[prsa_data.PRES <= 992] = 992
prsa_data.PRES[prsa_data.PRES >= 1034] = 1034
```

```python
sns.heatmap(prsa_data.isnull(), yticklabels=False, cbar=False)
```

```python
undef = prsa_data.isnull().sum(axis=1)
undef[undef >= 2]
```

```python
prsa_data = prsa_data.drop(undef[undef >= 2].index, axis=0)
```

```python
prsa_data.isnull().sum()
```

```python
prsa_data.SO2 = prsa_data.SO2.fillna(prsa_data.PRES.mean())
```

```python
filler = prsa_data.O3[prsa_data.O3.isna()]
```

```python
filler = prsa_data.O3[~prsa_data.O3.isna()].sample(n=len(filler)).set_axis(filler.index)
```

```python
prsa_data.O3 = prsa_data.O3.fillna(filler)
```

```python
prsa_data['O3'] = np.array(np.round((prsa_data['O3'])), dtype='int')
prsa_data.head()
```

```python
prsa_data.SO2 = np.log(prsa_data.SO2)
```

```python
quantitative_columns = ['SO2', 'NO2', 'CO', 'O3', 'PRES', 'WSPM']

# Устанавливаем стиль для графиков
sns.set(style="whitegrid")

# 1. Гистограммы для каждого атрибута
plt.figure(figsize=(15, 10))
for i, col in enumerate(quantitative_columns, 1):
    plt.subplot(3, 3, i)  # 3 строки, 3 столбца
    sns.histplot(prsa_data[col].dropna(), kde=False, bins=20, color='skyblue')
    plt.title(f'Гистограмма {col}')
    plt.xlabel(col)
    plt.ylabel('Частота')

plt.tight_layout()
plt.show()

# 2. Графики плотности для каждого атрибута
plt.figure(figsize=(15, 10))
for i, col in enumerate(quantitative_columns, 1):
    plt.subplot(3, 3, i)
    sns.kdeplot(prsa_data[col].dropna(), shade=True, color='orange')
    plt.title(f'График плотности {col}')
    plt.xlabel(col)
    plt.ylabel('Плотность')

plt.tight_layout()
plt.show()

# 3. Коробочные диаграммы для каждого атрибута
plt.figure(figsize=(15, 10))
for i, col in enumerate(quantitative_columns, 1):
    plt.subplot(3, 3, i)
    sns.boxplot(x=prsa_data[col].dropna(), color='lightgreen')
    plt.title(f'Коробочная диаграмма {col}')
    plt.xlabel(col)

plt.tight_layout()
plt.show()
```

```python
# 1. Столбчатая диаграмма для распределения классов чистоты воздуха (AQI Label)
plt.figure(figsize=(8, 6))
sns.countplot(x='AQI Label', data=prsa_data, palette='Set2')
plt.title('Распределение классов чистоты воздуха')
plt.xlabel('AQI Label')
plt.ylabel('Количество наблюдений')
plt.xticks(rotation=45)
plt.show()
```

```python
quantitative_columns = ['SO2', 'NO2', 'CO', 'O3', 'PRES', 'WSPM']

# Устанавливаем процентили для клиппинга
lower_percentile = 1  # Нижний процентиль (например, 1-й процентиль)
upper_percentile = 99  # Верхний процентиль (например, 99-й процентиль)

# Реализуем адаптивный клиппинг через процентили
for col in quantitative_columns:
    lower_limit = np.percentile(prsa_data[col].dropna(), lower_percentile)
    upper_limit = np.percentile(prsa_data[col].dropna(), upper_percentile)

    # Применяем клиппинг
    prsa_data[col] = prsa_data[col].clip(lower=lower_limit, upper=upper_limit)

# Печатаем обновленный DataFrame, чтобы увидеть изменения
print(prsa_data[quantitative_columns].describe())
```

```python
numerical_columns = ['SO2', 'NO2', 'CO', 'O3', 'PRES', 'WSPM']
categorical_columns = ['wd', 'AQI Label']  # Пример категориальных переменных

# Для числовых признаков заполняем пропуски медианой
for col in numerical_columns:
    prsa_data[col].fillna(prsa_data[col].median(), inplace=True)

# Для категориальных признаков заполняем пропуски наиболее частым значением (модой)
for col in categorical_columns:
    prsa_data[col].fillna(prsa_data[col].mode()[0], inplace=True)

# Проверим, остались ли пропуски
print(prsa_data.isnull().sum())
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler
# Список числовых признаков
numerical_columns = ['SO2', 'NO2', 'CO', 'O3', 'PRES', 'WSPM']

# 1. Применение Min-Max нормализации
min_max_columns = ['SO2', 'NO2', 'CO']  # Примеры столбцов, где Min-Max нормализация будет уместна

# 2. Применение стандартизации (Z-оценка)
standardization_columns = ['O3', 'PRES']  # Примеры столбцов, где применим стандартный метод (Z-оценка)

# 3. Применение логарифмической нормализации для сильно скошенных данных
log_transform_columns = ['WSPM']  # Пример скошенного признака

# Нормализуем с помощью MinMaxScaler
scaler_min_max = MinMaxScaler()
prsa_data[min_max_columns] = scaler_min_max.fit_transform(prsa_data[min_max_columns])

# Нормализуем с помощью StandardScaler
scaler_standard = StandardScaler()
prsa_data[standardization_columns] = scaler_standard.fit_transform(prsa_data[standardization_columns])

# Применим логарифмическое преобразование (добавляем 1, чтобы избежать ошибки с нулями)
prsa_data[log_transform_columns] = prsa_data[log_transform_columns].apply(lambda x: np.log1p(x))

# Проверим результат
print(prsa_data.head())
```

```python
# 1. Рассчитываем корреляцию между числовыми признаками
numerical_columns = ['SO2', 'NO2', 'CO', 'O3', 'PRES', 'WSPM']  # Список числовых признаков

correlation_matrix = prsa_data[numerical_columns].corr()

# 2. Строим тепловую карту
plt.figure(figsize=(10, 8))  # Размер графика
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', vmin=-1, vmax=1)
plt.title('Коррелограмма')
plt.show()

# 3. Анализируем значимость признаков
# Выводим корреляционную матрицу для более подробного анализа
print(correlation_matrix)
```
